```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Function to read convergence history data from file
def read_convergence_history(filename):
    iterations = []
    costs = []
    with open(filename, 'r') as file:
        for line in file:
            if line.strip():  # Skip empty lines
                iteration, cost = map(float, line.split())
                iterations.append(int(iteration))
                costs.append(cost)
    return iterations, costs

# Read the convergence history data
filename = 'convergence_history.txt'
iterations, costs = read_convergence_history(filename)

# Plot the convergence history
plt.figure(figsize=(10, 6))
plt.plot(iterations, costs, label='Cost')
plt.xlabel('Iteration')
plt.ylabel('Cost')
plt.title('Convergence History')
plt.legend()
```

```python
# Load training data
training_data = np.loadtxt("training_data_plot.dat")
X_train = training_data[:, :2]  # First two columns: inputs
y_train = training_data[:, 2]   # Last column: target outputs

# Load boundary data
boundary_data = np.loadtxt("boundary_data.txt")
X1, X2, outputs = boundary_data[:, 0], boundary_data[:, 1], boundary_data[:, 2]

# Reshape boundary data for contour plot
X1_grid = X1.reshape(100, 100)
X2_grid = X2.reshape(100, 100)
outputs_grid = outputs.reshape(100, 100)

# Plot decision boundary
plt.contour(X1_grid, X2_grid, outputs_grid, levels=[0.5], colors='black', linewidths=2)

# Plot training data
plt.scatter(X_train[y_train == -1, 0], X_train[y_train == -1, 1], color='red', label='Class 0')
plt.scatter(X_train[y_train == 1, 0], X_train[y_train == 1, 1], color='blue', label='Class 1')

plt.legend()
```
