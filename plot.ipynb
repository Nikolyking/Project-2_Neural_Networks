```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.lines import Line2D
```

```python
# Function to read convergence history data from file
def read_convergence_history(filename):
    iterations = []
    costs = []
    with open(filename, 'r') as file:
        for line in file:
            if line.strip():  # Skip empty lines
                iteration, cost = map(float, line.split())
                iterations.append(int(iteration))
                costs.append(cost)
    return iterations, costs
```

```python
### Training project

# Read the convergence history data
filename = 'convergence_history (2, 3, 3, 1)_shuffled.txt'
iterations, costs = read_convergence_history(filename)

# Plot the convergence history
plt.figure(figsize=(10, 6))
plt.plot(iterations, costs, label='Cost')
plt.xlabel('Iteration')
plt.ylabel('Cost')
plt.title('Convergence History')
plt.legend()
```

```python
iterations_1, costs_1 = read_convergence_history('convergence_history_simple_session_1.txt')
iterations_2, costs_2 = read_convergence_history('convergence_history_simple_session_2.txt')
iterations_3, costs_3 = read_convergence_history('convergence_history_simple_session_3.txt')
iterations_4, costs_4 = read_convergence_history('convergence_history_simple_session_4.txt')

plt.figure(figsize=(10, 6))
plt.plot(iterations_1, costs_1, label='Session 1')
plt.plot(iterations_2, costs_2, label='Session 2')
plt.plot(iterations_3, costs_3, label='Session 3')
plt.plot(iterations_4, costs_4, label='Session 4')
plt.xlabel('Iteration')
plt.ylabel('Cost')
plt.title('Convergence History for Multiple Sessions')
plt.legend()
```

```python
training_data = np.loadtxt("project_training_data_plotting.dat")

X_train = training_data[:, :2]
y_train = training_data[:, 2]

# Load boundary data
boundary_data_1 = np.loadtxt("boundary_data_simple_session_1.txt")
boundary_data_2 = np.loadtxt("boundary_data_simple_session_2.txt")
boundary_data_3 = np.loadtxt("boundary_data_simple_session_3.txt")
boundary_data_4 = np.loadtxt("boundary_data_simple_session_4.txt")

X1_1, X2_1, outputs_1 = boundary_data_1[:, 0], boundary_data_1[:, 1], boundary_data_1[:, 2]
X1_2, X2_2, outputs_2 = boundary_data_2[:, 0], boundary_data_2[:, 1], boundary_data_2[:, 2]
X1_3, X2_3, outputs_3 = boundary_data_3[:, 0], boundary_data_3[:, 1], boundary_data_3[:, 2]
X1_4, X2_4, outputs_4 = boundary_data_4[:, 0], boundary_data_4[:, 1], boundary_data_4[:, 2]

X1_grid_1 = X1_1.reshape(100, 100)
X2_grid_1 = X2_1.reshape(100, 100)
outputs_grid_1 = outputs_1.reshape(100, 100)

X1_grid_2 = X1_2.reshape(100, 100)
X2_grid_2 = X2_2.reshape(100, 100)
outputs_grid_2 = outputs_2.reshape(100, 100)

X1_grid_3 = X1_3.reshape(100, 100)
X2_grid_3 = X2_3.reshape(100, 100)
outputs_grid_3 = outputs_3.reshape(100, 100)

X1_grid_4 = X1_4.reshape(100, 100)
X2_grid_4 = X2_4.reshape(100, 100)
outputs_grid_4 = outputs_4.reshape(100, 100)

# Plot decision boundaries
plt.contour(X1_grid_1, X2_grid_1, outputs_grid_1, levels=[0.5], colors='blue', linewidths=2, label='Session 1')
plt.contour(X1_grid_2, X2_grid_2, outputs_grid_2, levels=[0.5], colors='orange', linewidths=2, label='Session 2')
plt.contour(X1_grid_3, X2_grid_3, outputs_grid_3, levels=[0.5], colors='red', linewidths=2, label='Session 3')
plt.contour(X1_grid_4, X2_grid_4, outputs_grid_4, levels=[0.5], colors='green', linewidths=2, label='Session 4')

# Plot training data
plt.scatter(X_train[y_train == -1, 0], X_train[y_train == -1, 1], color='blue', label='Blue')
plt.scatter(X_train[y_train == 1, 0], X_train[y_train == 1, 1], color='red', label='Red')

legend_elements = [
    Line2D([0], [0], color='blue', lw=2, label='Session 1'),
    Line2D([0], [0], color='orange', lw=2, label='Session 2'),
    Line2D([0], [0], color='green', lw=2, label='Session 3'),
    Line2D([0], [0], color='red', lw=2, label='Session 4')
]
plt.legend(handles=legend_elements, loc='upper right')
```
